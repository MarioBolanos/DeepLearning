#Two dimensional  linear systems




$$ \dot x=Ax $$

Where

\begin{equation}
A=
\begin{bmatrix}
a & b \\ c & d 
\end{bmatrix}
 \quad and \quad X=
 \begin{bmatrix}
x \\ y 
\end{bmatrix}
\end{equation}








```
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```
def BiDimentional (t,xy,alpha,beta,gamma,delta):
  x,y=xy
  return [x+y, 4*x-2*y]


a,b,c,d= 0.1, 0.02,0.3,0.01


t_in=0
t_fin=1
xy0=[0.1,4]
tspam=[t_in,t_fin]
t=np.linspace(t_in,t_fin,100)
sol=solve_ivp(BiDimentional,tspam,xy0,args=(a,b,c,d),t_eval=t)
plt.plot(sol.t, sol.y[0],label='prey')
plt.plot(sol.t, sol.y[1],label='predators')
#plt.axis([t_in,t_fin,0,15])
plt.legend()

#plt.axis([ti, tf, -10, 10])
plt.grid()
```

```
plt.plot(sol.y[0], sol.y[1],label='prey',color='blue')
plt.grid()
plt.ylabel('prey')
plt.xlabel('predator')
plt.show()

```

```
for i in np.arange(-5,5,1):
  for j in np.arange(-5,5,1):
    sol=solve_ivp(BiDimentional,tspam,[i,j],args=(a,b,c,d),t_eval=t)
    plt.plot(sol.y[0], sol.y[1])

#plt.axis([t_in,t_fin,0,15])

plt.axis([-5, 5, -5, 5])
plt.grid()
```
